```python
class grandparent():
  def family_name(self):
    print("Family Name: Hassan")
```

```python
class parent(grandparent):
  def occupation(self):
    print("Occupation: Data Scientists")
```

```python
class child(parent):
  def hobby(self):
    print("Hobby: Playing Cricket")
```

```python
c = child()
```

```python
c.family_name()
```

```python
c.occupation()
```

```python
c.hobby()
```

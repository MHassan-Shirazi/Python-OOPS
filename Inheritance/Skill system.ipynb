```python
class father():
  def skill(self):
    return "Scientists"
```

```python
class mother():
  def skill(self):
    return "Freelancer"
```

```python
class child(father , mother):
  def skill(self):
    return father.skill(self) + "and" + mother.skill(self)
```

```python
c = child()
```

```python
print("Child Skills:" , c.skill())
```

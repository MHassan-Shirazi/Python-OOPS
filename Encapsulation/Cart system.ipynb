```python
class shoppingcart():
  def __init__(self):
    self.__items = []

  def add_item(self , item):
    if item not in self.__items:
      self.__items.append(item)
      print(f"{item} has added")
    else:
      print(f"{item} is already in cart")


  def remove_item(self , item):
    if item in self.__items:
      self.__items.remove(item)
      print(f"{item} item has removed from cart")
    else:
       print(f"{item} item was not in card")


  def view_cart(self):
    if len(self.__items) == 0:
      print("cart is empty")
    else:
      print("Cart Items" , self.__items)
```

```python
cart = shoppingcart()
```

```python
cart.add_item("Apple")
```

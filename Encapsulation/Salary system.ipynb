```python
class Employee():
  def __init__(self , name , salary):
    self.__name = name
    self.__salary = salary

  def get_name(self):
    return self.__name

  def set_salary(self , salary):
    if salary > 0:
      self.__salary = salary
    else:

     print("Salary is negative")

  def show(self):
       print(f"Name: {self.__name} , Salary: {self.__salary}")
```

```python
emp = Employee("Hassan" , 100000)
```

```python
print(emp.get_name())
```

```python
emp.set_salary(200000)
```

```python
emp.show()
```
